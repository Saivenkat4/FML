```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import linear_model
```

```python
data =pd.read_csv("/content/Boston.csv")
data.head()
```

```python
data=pd.read_csv("/content/Boston.csv")
data.tail(-1)
```

```python
x=data.drop('lstat',axis=1).values
y=data.drop('lstat',axis=1).values
data.head()
```

```python
x_rm = x[:,5]
y = y[:, 0]
```

```python
type(x_rm),type(y)
```

```python
x_rm.shape,y.shape
```

```python
x_rm = x_rm.reshape(-1,1)
y=y.reshape(-1,1)
```

```python
x_rm.shape
```

```python
y.shape
```

```python
plt.scatter(x_rm,y)
plt.show()
```

```python
plt.scatter(x_rm,y)
plt.xlabel("no.of rows")
plt.ylabel("value of houses")
plt.show()
```

```python
#Fitting a regression model
reg = linear_model.LinearRegression()
reg.fit(x_rm, y)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
y_pred = reg.predict(x_rm)

print(f'MAE = {mean_absolute_error(y, y_pred)}')
print(f'MSE = {mean_squared_error(y, y_pred)}')
print(f'RMSE = {np.sqrt(mean_squared_error(y, y_pred))}')
print(f'R_2 = {r2_score(y, y_pred)}')
```

```python
prediction_space = np.linspace(min(x_rm),max(x_rm)).reshape(-1,1)
```

```python
plt.scatter(x_rm,y,color='purple')
plt.plot(x_rm,reg.predict(x_rm),color='blue',linewidth=3)
plt.show()
```

```python
plt.scatter(x_rm,y,color='red')
plt.plot(prediction_space,reg.predict(prediction_space),color='orange',linewidth=3)
plt.show()
```
